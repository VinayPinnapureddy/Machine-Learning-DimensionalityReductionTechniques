```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("mnist_train.csv")
#df.head()
labels=df["label"]
print(labels)
del df["label"]
df.head()
data=df.head(5000)
#data.head()
```

```python
labels=labels[0:5000]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
std_data=sc.fit_transform(data)
```

```python
from sklearn.manifold import TSNE
model=TSNE(n_components=2,random_state=0)
tsne_data=model.fit_transform(std_data)
```

```python
tsne_data=np.vstack((tsne_data.T,labels)).T
```

```python
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim1","Dim2","Labels"))
```

```python
sns.FacetGrid(tsne_df,hue="Labels",height=8).map(plt.scatter,"Dim1","Dim2").add_legend()
```

```python
from sklearn.manifold import TSNE
model=TSNE(n_components=2,random_state=0,perplexity=50)
tsne_data=model.fit_transform(std_data)
```

```python
tsne_data=np.vstack((tsne_data.T,labels)).T
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim1","Dim2","Labels"))
sns.FacetGrid(tsne_df,hue="Labels",height=8).map(plt.scatter,"Dim1","Dim2").add_legend()
```

```python
from sklearn.manifold import TSNE
model=TSNE(n_components=2,random_state=0,perplexity=50,n_iter=2500)
tsne_data=model.fit_transform(std_data)
```

```python
tsne_data=np.vstack((tsne_data.T,labels)).T
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim1","Dim2","Labels"))
sns.FacetGrid(tsne_df,hue="Labels",height=8).map(plt.scatter,"Dim1","Dim2").add_legend()
```
